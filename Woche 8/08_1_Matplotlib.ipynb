### 8.1.1. [*plot()*](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html)

Einfacher Plot:

```python
y = [3,2,7,5]

plt.plot(y)
plt.show()
```

Einfacher $x$-$y$-Plot:

```python
x = [-0.1,0.2,0.7,2.5]
y = [3,2,7,5]

plt.plot(x, y)
plt.show()
```

### 8.1.3. Plotgröße

Hierfür eine Abbildung mit [*figure()*](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html) erstellen.

`figsize` : (Breite, Höhe) in Zoll

```python
plt.figure(num=1,figsize=(15,3))
x = [-0.1,0.2,0.7,2.5]
y = [3,2,7,5]
plt.plot(x, y, 'ro:')
plt.show()
```

### 8.1.4. [*savefig()*](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.savefig.html)

Abspeichern von Plots

```python
plt.figure(2,(4,3))
x = [-0.1,0.2,0.7,2.5]
y = [3,2,7,5]
plt.plot(x, y, 'ro:')

plt.savefig('ein_plot.png') 
#plt.savefig('ein_plot.pdf') 
#plt.savefig('ein_plot.eps') 
#plt.savefig('ein_plot_hires.png',dpi=1000) 
```

Weißen Rand entfernen

```python
plt.figure(3,(4,3))
x = [-0.1,0.2,0.7,2.5]
y = [3,2,7,5]
plt.plot(x, y, 'ro:')

plt.savefig('ein_plot_ohne_rand1.png', bbox_inches='tight')
```

Weißen Rand entfernen bei gleichbleibender Abbildungsgröße

```python
plt.figure(4,(4,3))

x = [-0.1,0.2,0.7,2.5]
y = [3,2,7,5]
plt.plot(x, y, 'ro:')

plt.tight_layout()

plt.savefig('ein_plot_ohne_rand2.png')
```

```python
plt.figure(4,(4,3))

x = [-0.1,0.2,0.7,2.5]
y = [3,2,7,5]
plt.plot(x, y, 'ro:')

plt.subplots_adjust(left=0.05, right=0.99, top=1.0, bottom=0.075)

plt.savefig('ein_plot_ohne_rand3.png')
```

### 8.1.5. Subplots

[*subplot()*](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)

```python
x = [-0.1,0.2,0.7,2.5]
y = [3,2,7,5]

plt.figure(4,(8,5))

plt.subplot(2,3,1)
plt.plot(x, y, 'ro:')

plt.subplot(2,3,3)
plt.plot(x, y, 'og')

plt.subplot(234)
plt.plot(x, y, '^')

plt.subplot(235)
plt.plot(x, y, '--k')

plt.show()
```

[*subplots_adjust()*](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots_adjust.html)

```python
x = [-0.1,0.2,0.7,2.5]
y = [3,2,7,5]

plt.figure(4,(8,5))

plt.subplot(2,3,1)
plt.plot(x, y, 'ro:')

plt.subplot(2,3,3)
plt.plot(x, y, 'og')

plt.subplot(234)
plt.plot(x, y, '^')

plt.subplot(235)
plt.plot(x, y, '--k')

plt.subplots_adjust(left=0.05, right=0.99, top=1.0, bottom=0.075, wspace=0.6, hspace=0.01)
plt.show()
```

### 8.1.6. Achsen

Beschriftung

```python
x = [-0.1,0.2,0.7,2.5]
y = [3,2,7,5]

plt.plot(x, y, 'ro:')
plt.xlabel('$\\xi_{gesamt}$ / m')
plt.ylabel('T / K')
plt.show()
```

Limits

```python
x = [-0.1,0.2,0.7,2.5]
y = [3,2,7,5]

plt.plot(x, y, 'ro:')
plt.xlim(0.5,1.5)
plt.ylim([4,10])
plt.show()
```

Skala

```python
x = [-0.1,0.2,0.7,2.5]
y = [3,2,7,5]

plt.plot(x, y, 'ro:')
plt.xticks(np.linspace(-0.2,2.6,15))
plt.yticks(range(2,8),['A','','B','B 1/2','C'])
plt.grid(True)
plt.show()
```
